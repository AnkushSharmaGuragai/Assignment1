```python
# Calculate IQR and outlier thresholds for Score and GDP per Capita
score_iqr = south_asia_df['score'].quantile(0.75) - south_asia_df['score'].quantile(0.25)
gdp_iqr = south_asia_df['Log GDP per capita'].quantile(0.75) - south_asia_df['Log GDP per capita'].quantile(0.25)

score_bounds = [
    south_asia_df['score'].quantile(0.25) - 1.5 * score_iqr,
    south_asia_df['score'].quantile(0.75) + 1.5 * score_iqr
]
gdp_bounds = [
    south_asia_df['Log GDP per capita'].quantile(0.25) - 1.5 * gdp_iqr,
    south_asia_df['Log GDP per capita'].quantile(0.75) + 1.5 * gdp_iqr
]

# Identify all outliers
outliers = south_asia_df[
    (south_asia_df['score'] < score_bounds[0]) | (south_asia_df['score'] > score_bounds[1]) |
    (south_asia_df['Log GDP per capita'] < gdp_bounds[0]) | (south_asia_df['Log GDP per capita'] > gdp_bounds[1])
]

outliers[['Country name', 'score', 'Log GDP per capita']]
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=south_asia_df, x='Log GDP per capita', y='score', color='blue', label='Non-Outliers'
)
sns.scatterplot(
    data=outliers, x='Log GDP per capita', y='score', color='red', label='Outliers', s=100
)
plt.title('Scatter Plot of GDP per Capita vs. Score)')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.grid()
plt.show()
```

```python
# 6. Outlier Detection
# Define a function to detect outliers based on the IQR method
def detect_outliers(column, region_data):
    # Calculate the first (Q1) and third quartiles (Q3)
    Q1 = region_data[column].quantile(0.25)
    Q3 = region_data[column].quantile(0.75)

    # Compute the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Determine the lower and upper bounds for outlier detection
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return region_data[(region_data[column] < lower_bound) | (region_data[column] > upper_bound)]

# Apply the outlier detection function for both South Asia and Middle East regions
south_asia_outliers = detect_outliers('score', south_asia_df)
middle_east_outliers = detect_outliers('score', middle_east_df)

# Plotting the scatter plot and highlighting the outliers for both regions
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the scatter points for South Asia and Middle East, colored purple and yellow respectively
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_df, color='purple', label='South Asia')
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_df, color='yellow', label='Middle East')

# Highlight outliers in South Asia (colored pink) and Middle East (colored cyan)
sns.scatterplot(x=south_asia_outliers['Log GDP per capita'], y=south_asia_outliers['score'], color='pink', label='Outliers (South Asia)')
sns.scatterplot(x=middle_east_outliers['Log GDP per capita'], y=middle_east_outliers['score'], color='cyan', label='Outliers (Middle East)')

plt.title('Outliers in Score and GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Score')
plt.legend()

plt.show()


```
